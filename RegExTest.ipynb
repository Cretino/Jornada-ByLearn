# Trabalhando com RegEX
Pequeno exemplo para entender o funcionamento do regex.


```
# Primeiro importamos o pacote interno para poder utilizar as funções do RegEX.
import re
```

```
# Teste para validar números de celular com DDD.
# Qualquer número no intervalo de 0-9 com pelo menos 10 digitos e máximo de 11 digitos.
regex_expression = r'^[\d]{10,11}$'

# Variável para salvar o número do celular
# Definimos o valor da variável como '1' para que a aplicação execute pelo menos uma vez.
phone_number = '1'
regex_result = None

while phone_number != '0' and regex_result == None:
  phone_number = input('Digite seu número do celular: (Digite \'0\' para encerrar a aplicação.)\n')
  regex_result = re.search(regex_expression, phone_number)

if phone_number == '0':
  print('Tudo bem!')
elif regex_result != None:
  print(f'Seu número de celular é: {phone_number}')
```

```
# Teste para validar nome completo.
# Pelo menos 4 letras entre a-z (Maiúsculo ou minúsculo) com 0~2 ocorrências de espaço seguidos de pelo menos uma letra entre a-z. (Maiúsculo ou minúsculo)
regex_expression = r'^[a-zA-Z]{4,}(?: [a-zA-Z]+){0,2}$'

# Variável para salvar o nome
# Definimos o valor da variável como '1' para que a aplicação execute pelo menos uma vez.
full_name = '1'
regex_result = None

while full_name != '0' and regex_result == None:
  full_name = input('Digite seu nome completo: (Digite \'0\' para encerrar a aplicação.)\n')
  regex_result = re.search(regex_expression, full_name)

  if full_name == '0':
    print('Tudo bem.')
  elif regex_result != None:
    print(f'O seu nome é: {full_name}')
```

```
# Teste para validar data de nascimento.
# Formato (00/00/0000 ou 0/0/0000)
regex_expression = r'^[\d]{1,2}\/[\d]{1,2}\/[\d]{4}$'

# Variável para salvar a data
# Definimos o valor da variável como '1' para que a aplicação execute pelo menos uma vez.
birth_date = '1'
regex_result = None

while birth_date != '0' and regex_result == None:
  birth_date = input('Digite sua Data de Nascimento: (Digite \'0\' para encerrar a aplicação. Formato: dd/mm/aaaa\n')
  regex_result = re.search(regex_expression, birth_date)

  if birth_date == '0':
    print('Tudo bem.')
  elif regex_result != None:
    print(f'Sua data de nascimento é: {birth_date}')
```
